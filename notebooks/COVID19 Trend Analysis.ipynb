```python
df2 = df.groupby('Country')[['Confirmed','Deaths','Recovered']].sum().reset_index()
```

```python
df2
```

```python
df2 = df.groupby(['Country','Date'])[['Confirmed','Deaths','Recovered']].sum().reset_index()
```

```python
df2
```

```python
df3 = df2[df2['Confirmed']>10000]
```

```python
df3
```

```python
countries = ['US','India','Brazil','Russia','Iran','Iraq','Israel']
len(countries)
```

```python
for index in range(0,len(countries)):
    C = df3[df3['Country']==countries[index]].reset_index()
    plt.figure(figsize=(5,5))
    plt.plot(np.arange(0,len(C)),C['Confirmed'],color='orange',label='Confirmed')
    plt.plot(np.arange(0,len(C)),C['Deaths'],color='red',label='Deaths')
    plt.plot(np.arange(0,len(C)),C['Recovered'],color='green',label='Recovered')
    plt.xlabel('Days since first case')
    plt.ylabel('Number of Cases')
    plt.title(countries[index])
    plt.grid(True, linestyle='--', alpha=0.4)
    plt.legend()
    plt.show()
```

```python
df4 = df3.groupby(['Date'])[['Confirmed','Deaths','Recovered']].sum().reset_index()
```

```python
C = df4
plt.scatter(np.arange(0,len(C)),C['Confirmed'],color='orange',label='Confirmed')
plt.scatter(np.arange(0,len(C)),C['Deaths'],color='red',label='Deaths')
plt.scatter(np.arange(0,len(C)),C['Recovered'],color='green',label='Recovered')
plt.ticklabel_format(style='plain', axis='y')
plt.grid(True, linestyle='--', alpha=0.2)
plt.title('Global COVID-19 Trend Analysis')
plt.xlabel('Days Since First Recorded Case')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```

```python
C = df4

plt.figure(figsize=(12,6))

plt.plot(C['Date'], C['Confirmed'], color='orange', label='Confirmed')
plt.plot(C['Date'], C['Deaths'], color='red', label='Deaths')
plt.plot(C['Date'], C['Recovered'], color='green', label='Recovered')

plt.ticklabel_format(style='plain', axis='y')
plt.yscale('log')
plt.grid(True, linestyle='--', alpha=0.5)

plt.title('Global COVID-19 Trend Analysis')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Number of Cases')

plt.legend()
plt.show()
```
